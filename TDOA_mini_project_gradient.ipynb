```python
import numpy as np
import math
sensor_pos=np.array([(50,120),(52.91,-60),(0,50),(20,77.45)])
time=np.array([13,8,5,8])
c=10



err=np.zeros((301,301))
for i in range(0,301,1):
    for j in range(0,301,1):
        for k in range(0,4,1):
            err[i][j] += abs( ( math.sqrt( ((i-150)-sensor_pos[k][0])**2 + ((j-150)-sensor_pos[k][1])**2 ) ) / c - time[k] )
            
min_err=np.amin(err)

print(min_err)

print(np.where(err==np.amin(err)))


# the min_err is quite low and the point where min occurs has index [1150][150] representing the origin
```

```python
import matplotlib.pyplot as plt

x=np.zeros((301))
y=np.zeros((301))

for alpha in range(0,301,1):
    x[alpha]=alpha

for beta in range(0,301,1):
    y[beta]=beta
    
plt.contourf(x,y,err, 420)


plt.plot()
```
